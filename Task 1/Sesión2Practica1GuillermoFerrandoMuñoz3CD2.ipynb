Construir un tokenizador para el espaÃ±ol, que, dado un fichero de texto de
entrada (entrada_tokenizador.txt), separe en tokens, y los muestre en un
fichero de salida en el formato que se muestra en (salida_tokenizador.txt).
Por lo menos el tokenizador deberÃ¡ funcionar correctamente para el ejemplo.
El tokenizador debe cumplir las siguientes restricciones:

```python
#1) Los sÃ­mbolos que hay que separar de cada palabra son: ( ) . , â€˜ â€œ ? Â¿ ! Â¡ ... ; :
p1=r"\.{3}|[().,'â€œ?Â¿!Â¡;:]"

#2) No se deben separar los nÃºmeros decimales, ejemplo: 44,45 45.60
p2=r"\d+\.\d+|\d+,\d+"

#3) No se deben separar fechas 12/12/90, 12-03-98, ni horas, 9:30 h.
p3=r"\d+-\d+-?\d*|\d+/\d+/?\d*|\d+:\d+\s+\w+"

#4) Las fechas en formato 12 de febrero de 2018 hay que mantenerlas como un token
p4=r"\d+\s\w+\s\w+\s\w+\d+"

#5) No se deben separar direcciones web http://www.colorin.com ni correos electrÃ³nicos xx@cdit.com
p5=r"\w+@\w+\.\w+\.?\w*|http\S+"

#6) Hay que mantener menciones a usuarios (@user) y hashtags (#hashtag) como se utilizan en Twitter.
p6=r"\s[@|#]\S+"

#7) Hay que mantener acrÃ³nimos, por ejemplo: EE.UU., S.L., CC.OO., S.A., D., U.R.S.S, ...
p7=r"[A-Z]+\.\S*"

#8) Respetar los emoticonos: 
p8=r"[\U00010000-\U0010ffff]"

#palabras
p9=r"\w+-?\w*"
```

Una vez tenemos las restricciones listas, podemos escribir el fichero de salida del texto tokenizado:

```python
import re

with open("entrada_tokenizador.txt", "r", encoding='UTF-8') as entrada:
    with open('salida_tokenizador.txt', 'w', encoding="UTF-8") as salida:
        for l in entrada:
            salida.write(l+'\n')
            print(l)
            tokenizador = re.findall(f'{p1}|{p2}|{p3}|{p4}|{p5}|{p6}|{p7}|{p8}|{p9}', l)
            for i in tokenizador:
                if len(i)>1 and ("@"==i[1] or "#"==i[1]):
                    i=i[1:]
                salida.write(i+'\n')
                print(i)
```
