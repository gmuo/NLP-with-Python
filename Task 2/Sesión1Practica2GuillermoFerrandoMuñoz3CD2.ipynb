```python
from whoosh.index import create_in, open_dir
from whoosh.fields import Schema, TEXT, KEYWORD, DATETIME, ID
import json
import os
```

Ejercicio 1:

Escribir un programa que indexe la colección proporcionada. Recibirá como
primer argumento el directorio donde están los documentos que se deben
indexar y como segundo argumento el directorio donde se guardará el índice.

```python
def indexador(direc1, direc2):
    schema = Schema(Id=ID(stored=True), title=TEXT(stored=True), 
                    article=TEXT,keywords=KEYWORD(stored=True), date=DATETIME(stored=True))
    index = create_in(direc2, schema)
    writer = index.writer()
    for dirpath, dirnames, filenames in os.walk(direc1):
        #print(dirpath, dirnames, filenames)
        for file in filenames:
            #print(file)
            if file.endswith(".json"):
                completo = os.path.join(dirpath, file)
                print(completo)
                with open(completo) as j:
                    JASON = json.load(j)
                    #print(JASON)
                    writer.add_document(Id=file, title=JASON["title"], 
                                        article=JASON["article"], keywords=JASON["keywords"], date=JASON["date"])
    writer.commit()
    
indexador("2018_2","S1_P2")
```
