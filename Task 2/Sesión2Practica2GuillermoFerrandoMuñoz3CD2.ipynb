```python
from whoosh.index import open_dir
from whoosh.qparser import QueryParser
import json
import os
```

```python
def load_json(filename):
    with open(filename) as fh:
        d = json.load(fh)
        return d
def save_dict_as_json(d, filename):
    with open(filename, "w") as fh:
        obj = json.dump(d, fh)
        
def saca_contenido():
    notis=[]
    for dirpath, dirnames, filenames in os.walk("2018_2"):
        #print(dirpath, dirnames, filenames)
        for file in filenames:
            #print(file)
            if file.endswith(".json"):
                completo = os.path.join(dirpath, file)
                #print(completo)
                notis.append(load_json(completo))
    return notis
```

Implementamos la función: `q` será `False` por defecto, si no, será el query a buscar desde la línea de comandos. Si es `False`, la función entrará en bucle y no parará hasta que el query introducido por teclado sea una cadena vacía. Por otra parte, `extend` podrá ser `True` si se quiere extender el resultado mostrado por pantalla, o `False` si no se quiere extender, siendo el valor por defecto:

```python
def SAR_p5_Searcher(direc, q=False, extend=False):
    ix = open_dir(direc)
    # con with n o es necesario cerrar el buscador
    with ix.searcher() as searcher:
        if q is not False:  # q es la cadena de búsqueda
            parser = QueryParser("title", ix.schema)
            query = parser.parse(q)
            r = searcher.search(query)
            #long=len([n for n,r in enumerate(r)])
            print(f'----------{len(r)} noticias recuperadas---------')
            for n,r in enumerate(r):
                try:
                    if extend is False: # no usamos extend
                        print(f"Noticia {n+1}: ")
                        print()
                        print('**Título**: ', r["title"])
                        print('**Fecha de publicación**: ', r["date"])
                        print('**Palabras-clave**: ', r["keywords"])
                        print()
                        print()
                    else: # usamos extend (extend = True)
                        print(f"Noticia {n+1}: ")
                        print()
                        print('**Título**: ', r["title"])
                        print('**Fecha de publicación**: ', r["date"])
                        print('**Palabras-clave**: ', r["keywords"])
                        for i in saca_contenido():
                            if r["title"]==i["title"] and r["date"]==i["date"]:
                                print('**Cuerpo completo**: ', i["article"])
                                break
                        print()
                        print()
                except IndexError:
                    pass
                    
        else: # q es False
            q2 = str(input('Escribe tu query (cadena vacía para finalizar): '))
            while q2 != " ":
                parser = QueryParser("title", ix.schema)
                query = parser.parse(q2)
                r = searcher.search(query)
                #long=len([n for n,r in enumerate(r)])
                print(f'----------{len(r)} noticias recuperadas---------')
                for n,r in enumerate(r):
                    try:
                        if extend is False: # no usamos extend
                            print(f"Noticia {n+1}: ")
                            print()
                            print('**Título**: ', r["title"])
                            print('**Fecha de publicación**: ', r["date"])
                            print('**Palabras-clave**: ', r["keywords"])
                            print()
                            print()
                        else: # usamos extend (extend = True)
                            print(f"Noticia {n+1}: ")
                            print()
                            print('**Título**: ', r["title"])
                            print('**Fecha de publicación**: ', r["date"])
                            print('**Palabras-clave**: ', r["keywords"])
                            for i in saca_contenido():
                                if r["title"]==i["title"] and r["date"]==i["date"]:
                                    print('**Cuerpo completo**: ', i["article"])
                                    break
                            print()
                            print()
                    except IndexError:
                        pass
                q2 = str(input('Escribe tu query (cadena vacía para finalizar): '))
                    
with open()
SAR_p5_Searcher("S1_P2", q=False, extend=False)
```
