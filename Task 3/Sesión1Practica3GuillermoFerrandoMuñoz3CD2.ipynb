
Utilizando el modelo `es_core_news_sm.load()`, eliminaremos aquellos tokens que sean declarados como `'PUNCT'` y `'SPACE'`. Algunos declarados como `'SYM'` se podrían eliminar también, pero no todos, por lo que no los eliminaremos. Utilizaremos la librería nltk para obtener una lista de stopwords.

````python
nltk.download('stopwords')
palabrasparadoras = stopwords.words('spanish') # lista de stopwords del español
print(palabrasparadoras)
````

````python
sentences=df_detests["sentence"].to_list()

pals={}
for n,s in enumerate(sentences):
    doc = nlp(s)
    pals[n] = []
    for d in doc:
        if d.text.lower() not in palabrasparadoras and (d.pos_!='PUNCT' and d.pos_!='SPACE'):
            pals[n].append(d.text)
            #print(d.text,  d.pos_, d.lemma_)
            
print(len(pals))
print(pals) # diccionario con oraciones (en forma de listas) con palabras filtradas
````

````python
joinpals = pals.copy()

for i in joinpals:
    joinpals[i]=" ".join(joinpals[i])

print(joinpals) # diccionario con oraciones con palabras filtradas
````

### 2. Word Embeddings

Necesitaremos la librería gensim para entrenar nuestro modelo de word embedding.

````python
#!pip install gensim
import numpy
import scipy
from gensim.models import Word2Vec, KeyedVectors
````

````python
#help(Word2Vec)
````

Con la función Word2Vec, entrenaremos el modelo. Ejemplo de uso de Word2Vec:

```python
from gensim.models import Word2Vec
sentences = [["cat", "say", "meow"], ["dog", "say", "woof"]]
model = Word2Vec(min_count=1)
model.build_vocab(sentences)  # prepare the model vocabulary
model.train(sentences, total_examples=model.corpus_count, epochs=model.epochs)  # train word vectors
```

Por defecto, los vectores de cada palabra tendrán 100 componentes, lo dejaremos así, ya que lo considero un buen número. Indicaremos que el mínimo de repetición de palabras sea 1, estadísticamente sería como mínimo "necesario" 5, pero las oraciones no son tan largas como para poder permitirnos eso.

````python
embedd = Word2Vec(min_count = 1)
embedd.build_vocab(pals.values())
embedd.train(pals.values(), total_examples=embedd.corpus_count, epochs=embedd.epochs) #modelo entrenado
````

Una vez entrenado el modelo, para cada oración, obtendremos el vector de cada palabra y calcularemos la media de la oración.

````python
#embedd.wv.get_vector("pasan")
vectors = {}
for n,l in pals.items():
    vectors[n] = [embedd.wv.get_vector(word) for word in l]
#print(vectors)
````

Hacemos un dataframe con el vector de cada palabra:

````python
nombresfilas = [pal for l in pals.values() for pal in l]
dfvectors = [v for l in vectors.values() for v in l]
pd.DataFrame(dfvectors, index = nombresfilas).head(10)
````

Calculamos ahora la media de cada oración:

````python
meanvectors={}

for oracion in vectors:
    try:
        meanvectors[oracion]=(sum(vector for vector in vectors[oracion]))/len(vectors[oracion])
        
    except:
        meanvectors[oracion]=numpy.zeros(100)
        
#print(meanvectors)
````

````python
res = pd.DataFrame(meanvectors.values()) # los nombres de las filas son la posición de las oraciones
res.head(10)
````
